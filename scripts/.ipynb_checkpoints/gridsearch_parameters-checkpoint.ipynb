## MCMC

```julia
### da lanciare dentro Genie

using DataFrames, DCAUtils, DelimitedFiles, Genie, Statistics, DataFrames, CSV

wt_name = "DHFR"

en_files = filter(f -> startswith(basename(f), "en"), 
    readdir("../simulation_data/$(wt_name)/mcmc/"; join=true)) ;
ham_files = filter(f -> startswith(basename(f), "ham") && !startswith(basename(f), "pair_ham"), 
    readdir("../simulation_data/$(wt_name)/mcmc/"; join=true));
pair_ham_files = filter(f -> startswith(basename(f), "ham") && !startswith(basename(f), "pair_ham"), 
    readdir("../simulation_data/$(wt_name)/mcmc/"; join=true));
msa_files = filter(f -> startswith(basename(f), wt_name*"_mcmc"), 
    readdir("../simulation_data/$(wt_name)/mcmc/"; join=true));

[en_files ham_files pair_ham_files msa_files]

if wt_name == "TEM"
    target_msa = read_fasta_alignment("../data/tem_round12_noflankgaps_nodupl.fa",0.9)
    wt = join(readdlm("../data/TEM1_pfam_DNA.fasta", skipstart = 1));
    L_big = round(Int,length(wt)/3)
    wt_DNA_seq = [wt[((i-1)*3 +1):(i*3)] for i in 3:L_big-1];
    wt_seq = [cod2amino[x] for x in wt_DNA_seq];
elseif wt_name == "PSE"
    target_msa = read_fasta_alignment("../data/pse_round20_noflankgaps_nodupl.fa",0.9)
    wt = join(readdlm("../data/PSE1_pfam_DNA.fasta", skipstart = 1));
    L_big = round(Int,length(wt)/3)
    wt_DNA_seq = [wt[((i-1)*3 +1):(i*3)] for i in 3:L_big-1];
    wt_seq = [cod2amino[x] for x in wt_DNA_seq];
elseif wt_name == "AAC"
    target_msa = read_fasta_alignment("../data/aac_round8_noflankgaps_nodupl.fa",0.9)
    wt = join(readdlm("../data/AAC6_pfam_DNA.fasta", skipstart = 1));
    L_big = round(Int,length(wt)/3)
    wt_DNA_seq = [wt[((i-1)*3 +1):(i*3)] for i in 1:L_big];
    wt_seq = [cod2amino[x] for x in wt_DNA_seq];
elseif wt_name == "DHFR"
    target_msa = read_fasta_alignment("../data/dhfr_round15_noflankgaps_nodupl.fa",0.9)
    wt = join(readdlm("../data/mDHFR_pfam_DNA.fasta", skipstart = 1));
    L_big = round(Int,length(wt)/3)
    wt_DNA_seq = [wt[((i-1)*3 +1):(i*3)] for i in 1:L_big];
    wt_seq = [cod2amino[x] for x in wt_DNA_seq];
else
    println("Error, specify correct wt_name")
end


# Read model parameters
println("Reading model parameters")
if wt_name == "AAC"
    file_model = "../data/Parameters_conv_AAC6.dat"
    h_tmp, J_tmp = read_par_BM_0gapsave(file_model)
elseif wt_name == "DHFR"
    file_model = "../data/Parameters_conv_DHFR.dat"
    h_tmp, J_tmp = read_par_BM_0gapsave(file_model)
elseif wt_name == "PSE" || wt_name == "TEM"
    file_model = "../data/model_natural_PSE.dat"
    h_tmp, J_tmp = read_par_BM_lettersave(file_model)
else
    println("Error, specify correct wt_name")
end
    
h = set_max_field_to_0save(h_tmp);
J_tmp2 = symmetrize_Jsave(J_tmp);
J = permutedims(J_tmp2, [1,3,2,4]);


en = energy(target_msa, h, J)
ham = ham_dist(wt_seq, target_msa)
@time pair_ham = pairwise_ham_dist(target_msa, n_seq = 1000, all = true)

target_en = mean(en)[1]
target_ham = mean(ham)[1]
target_pair_ham = mean(pair_ham)[1]

score_en = [abs((mean(readdlm(en_files[i]))[1] - target_en) /target_en) for i in 1:length(en_files)];
score_ham = [abs((mean(readdlm(ham_files[i]))[1] - target_ham) /target_ham) for i in 1:length(ham_files)];
score_pair_ham = [abs((mean(readdlm(pair_ham_files[i]))[1] - target_pair_ham) /target_pair_ham) 
    for i in 1:length(pair_ham_files)];

# Example: assume msa_files is already defined
# We'll extract the T and steps using a regex pattern

results = [match(r"T(?<T>[\d.]+)_steps(?<steps>\d+)", file) for file in msa_files]

# Extract values and clean file names
temps = [parse(Float64, r["T"]) for r in results if r !== nothing]
steps = [parse(Int, r["steps"]) for r in results if r !== nothing]
filenames = [basename(file) for (file, r) in zip(msa_files, results) if r !== nothing]


# Create the DataFrame
df = DataFrame(
    id = filenames,
    T = temps,
    steps = steps, 
    score_en = score_en,
    score_ham = score_ham,
    score_tot = score_ham .+ score_en,
    score_pair_ham = score_pair_ham
)

df[sortperm(df.score_tot),2:end]


CSV.write("$(wt_name)_mcmc.csv", df)


```

```julia
using CSV, DataFrames, PyPlot

function plot_heatmap_mcmc(filename::AbstractString; savepath::AbstractString="")
    # Carica i dati
    df = CSV.read(filename, DataFrame)

    # Estrai e ordina i valori unici
    temps = sort(unique(df.T))
    steps = sort(unique(df.steps))

    # Crea matrice NaN con righe = steps, colonne = temps
    heat = fill(NaN, length(steps), length(temps))

    # Riempi la matrice heat
    for row in eachrow(df)
        i = findfirst(==(row.steps), steps)
        j = findfirst(==(row.T), temps)
        heat[i, j] = row.score_tot / 2
    end

    # Trova valore minimo e posizione
    min_val = minimum(skipmissing(vec(heat)))
    min_pos = findfirst(x -> x == min_val, heat)
    min_i, min_j = min_pos[1], min_pos[2]

    # Crea figura e disegna heatmap
    figure(figsize=(8,6))
    imshow(heat; cmap="viridis", origin="lower", aspect="auto")

    # Imposta etichette
    xticks(0:length(temps)-1, string.(temps), rotation=45)
    yticks(0:length(steps)-1, string.(steps))
    xlabel("Temperature", fontsize = 30)
    ylabel("Steps", fontsize = 30)
    cb = colorbar()
    cb.set_label("% err on Energy and Hamming", fontsize=20)

    # Aggiungi testo minimo sulla cella corrispondente
    text(min_j-1, min_i-1, string(round(min_val, digits=3)),
         ha="center", va="center", color="white", fontsize=12, fontweight="bold")

    tight_layout()

    # Salva il file se savepath è specificato
    if savepath != ""
        savefig(savepath)
        println("Plot salvato in: $savepath")
    end

    # Mostra la figura (opzionale)
    # show()
end
```

```julia
using CSV, DataFrames, PyPlot
using FilePathsBase: basename, dirname, splitext, joinpath

function plot_heatmap_popgen(filename::AbstractString)
    # Load data
    df = CSV.read(filename, DataFrame)

    # Extract unique values
    temps = sort(unique(df.T))
    mu_binds = sort(unique(df.mu_bind))
    mus = sort(unique(df.mu))

    # Output path
    base = splitext(basename(filename))[1]
    savepath = joinpath(dirname(filename), "$(base)_panel_by_mu.png")

    # Prepare figure layout
    n = length(mus)
    ncols = ceil(Int, sqrt(n))
    nrows = ceil(Int, n / ncols)

    fig, axs = subplots(nrows, ncols, figsize=(4ncols, 4nrows), squeeze=false)

    # Variables to track global minimum
    global_min_val = Inf
    global_min_indices = nothing  # (mu_idx, i, j)

    # Store heatmaps to reuse (optional)
    heats = Vector{Matrix{Float64}}(undef, n)

    # First loop: build all heats and find global minimum
    for (k, mu_val) in enumerate(mus)
        # Filter data for current mu
        subdf = filter(row -> row.mu == mu_val, df)

        # Fill heatmap: rows = mu_bind, cols = T
        heat = fill(NaN, length(mu_binds), length(temps))
        for row in eachrow(subdf)
            i = findfirst(==(row.mu_bind), mu_binds)
            j = findfirst(==(row.T), temps)
            heat[i, j] = row.score_tot / 2
        end

        heats[k] = heat  # store
        


        if all(ismissing, heat)
            continue
        end

        local_min_val = minimum(x for x in heat if x !== missing && !isnan(x))
        #local_min_val = minimum(skipmissing(vec(heat)))
        local_min_pos = findfirst(x -> x == local_min_val, heat)

        if local_min_val < global_min_val
            global_min_val = local_min_val
            global_min_indices = (k, Tuple(local_min_pos))
        end
        #println("Mu is $(mu_val)")
        #println(heat)
        #println(local_min_val)
    end

    # Second loop: plot all heats, annotate only the global minimum
    for (k, mu_val) in enumerate(mus)
        row_idx = div(k - 1, ncols) + 1
        col_idx = mod(k - 1, ncols) + 1
        ax = axs[row_idx, col_idx]

        heat = heats[k]
        if heat === nothing || all(ismissing, heat)
            ax.axis("off")
            continue
        end

        im = ax.imshow(heat; cmap="viridis", origin="lower", aspect="auto")
        ax.set_title("μ = $mu_val", fontsize=15)
        ax.set_xticks(0:length(temps)-1)
        ax.set_xticklabels(string.(temps), rotation=45, fontsize=8)
        ax.set_yticks(0:length(mu_binds)-1)
        ax.set_yticklabels(string.(mu_binds), fontsize=8)

        # Annotate only if this subplot has the global minimum
        if global_min_indices !== nothing && global_min_indices[1] == k
            min_i, min_j = global_min_indices[2]
            ax.text(min_j-1, min_i-1, string(round(global_min_val, digits=3));
                    ha="center", va="center", color="white", fontsize=9, fontweight="bold")
        end
    end

    fig.text(0.5, 0.04, "T", ha="center", va="center", fontsize=28)
    fig.text(0.04, 0.5, "μ_bind", ha="center", va="center", rotation="vertical", fontsize=28)

    # Remove unused axes
    for i in 1:nrows, j in 1:ncols
        k = (i-1)*ncols + j
        if k > n
            axs[i, j].axis("off")
        end
    end

    savefig(savepath, bbox_inches="tight")
    println("✔️  Saved figure at: $savepath")
end
```

```julia
function plot_heatmap_mcmctree(filename::AbstractString; savepath::AbstractString="")
    # Carica i dati
    df = CSV.read(filename, DataFrame)

    # Estrai e ordina i valori unici
    temps = sort(unique(df.T))
    steps = sort(unique(df.steps))

    # Crea matrice NaN con righe = steps, colonne = temps
    heat = fill(NaN, length(steps), length(temps))

    # Riempi la matrice heat
    for row in eachrow(df)
        i = findfirst(==(row.steps), steps)
        j = findfirst(==(row.T), temps)
        heat[i, j] = row.score_tot / 2
    end

    # Filtra solo valori validi per minimo
    valid_vals = filter(!isnan, vec(heat))
    if isempty(valid_vals)
        println("Attenzione: nessun dato valido da plottare.")
        return
    end

    # Trova valore minimo e posizione
    min_val = minimum(valid_vals)
    # findfirst può fallire se ci sono NaN: qui cerco direttamente nel heat
    min_pos = findfirst(x -> x == min_val, heat)

    # Crea figura e disegna heatmap
    figure(figsize=(8,6))
    imshow(heat; cmap="viridis", origin="lower", aspect="auto", interpolation="none")

    # Imposta etichette
    xticks(0:length(temps)-1, string.(temps), rotation=45)
    yticks(0:length(steps)-1, string.(steps))
    xlabel("Temperature", fontsize = 30)
    ylabel("Steps", fontsize = 30)
    cb = colorbar()
cb.set_label("% err on Energy and Hamming", fontsize=20)

    # Aggiungi testo minimo solo se posizione trovata
    if min_pos !== nothing
        min_i, min_j = Tuple(min_pos)
        text(min_j-1, min_i-1, string(round(min_val, digits=3)),
             ha="center", va="center", color="white", fontsize=12, fontweight="bold")
    end

    tight_layout()

    # Salva il file se savepath è specificato
    if savepath != ""
        savefig(savepath)
        println("Plot salvato in: $savepath")
    end
end
```
