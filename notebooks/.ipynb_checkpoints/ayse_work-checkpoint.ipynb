## Trying to do Ayse work for 3 DBD sequences

```julia
using Revise, Genie, DCAUtils, JLD2, PyPlot, Statistics


nat_msa  = read_fasta_alignment("../Gen.jl/data/alignments/natural/DBD_alignment.uniref90.cov80.a2m", 0.9);
w = compute_weights(nat_msa, 22, 0.2)[1];
@load "../data_Genie/pars_dbd.jld2"; h = h_dbd; J = J_dbd;
@load"../data_Anc/3seqs_data_for_tree_sampling.jld2";

start_ens = energy(start_msa, h, J);

q = 21; L =size(nat_msa,1);

#just one
temp_AW = 0.2; temp_HI = 0.5; temp_LO = 1.0; res_all_AW = []; res_all_HI = []; res_all_LO = [];
N_steps_AW = 500; N_steps_HI = 500; N_steps_LO = 500; each_step = 50; N_seqs = 1000; p = 0.; 

for i in 1:3
    @time res_AW = run_evolution(hcat([start_msa[:,i] for _ in 1:N_seqs]...), h, J, temp = temp_AW,
    N_steps = N_steps_AW, each_step = each_step, p = p);
    push!(res_all_AW, res_AW)
    init = res_AW.step_msa[end]
    @time res_HI = run_evolution(init, h, J, temp = temp_HI,
    N_steps = N_steps_HI, each_step = each_step, p = p);
    push!(res_all_HI, res_HI)
    @time res_LO = run_evolution(init, h, J, temp = temp_LO,
    N_steps = N_steps_LO, each_step = each_step, p = p);
    push!(res_all_LO, res_LO)
end
    

N_points_AW = length(res_all_AW[1].step_msa); 
N_points_LO = length(res_all_LO[1].step_msa); 
N_points_HI = length(res_all_HI[1].step_msa); 

en_AW = []; var_en_AW = []; en_HI = []; var_en_HI = []; en_LO = []; var_en_LO = [];
for i in 1:3
    en = []; var_en = [];
    for t in 1:N_points_AW
        msa_en = energy(res_all_AW[i].step_msa[t], h, J) .- start_ens[i]
        push!(en, mean(msa_en)[1])
        push!(var_en, var(msa_en)[1])
    end
    push!(en_AW, en)
    push!(var_en_AW, var_en)
    en = []; var_en = [];
    for t in 1:N_points_HI
        msa_en = energy(res_all_HI[i].step_msa[t], h, J) .- start_ens[i]
        push!(en, mean(msa_en)[1])
        push!(var_en, var(msa_en)[1])
    end
    push!(en_HI, en)
    push!(var_en_HI, var_en)
    en = []; var_en = [];
    for t in 1:N_points_LO
        msa_en = energy(res_all_LO[i].step_msa[t], h, J) .- start_ens[i]
        push!(en, mean(msa_en)[1])
        push!(var_en, var(msa_en)[1])
    end
    push!(en_LO, en)
    push!(var_en_LO, var_en)
end


steps_LO = res_all_LO[1].steps ./ L; steps_HI = res_all_HI[1].steps ./ L; steps_AW = (2 .- reverse(res_all_AW[1].steps)) ./ L;
for i in 1:3
close("all"); plt.errorbar(steps_AW, .- en_AW[i], yerr = sqrt.(var_en_AW[i]), marker = "o", label = "AW (T = $(temp_AW))");
    plt.errorbar(steps_LO, .- en_LO[i], yerr = sqrt.(var_en_LO[i]), marker = "o", label = "LO (T = $(temp_LO))");
    plt.errorbar(steps_HI, .- en_HI[i], yerr = sqrt.(var_en_HI[i]), marker = "o", label = "HI (T = $(temp_HI))");
    plt.xlabel("MC Sweeps"); plt.ylabel(" - Delta Energy"); plt.legend(loc = "upper left"); 
    plt.title("SEQ $(i)"); plt.savefig("../figures/DBD_ayse_trial/en_SEQ_$(i).png") end

for i in 1:3
close("all"); plt.plot(steps_AW, var_en_AW[i], marker = "o", label = "AW (T = $(temp_AW))");
    plt.plot(steps_LO, var_en_LO[i], marker = "o", label = "LO (T = $(temp_LO))");
    plt.plot(steps_HI, var_en_HI[i], marker = "o", label = "HI (T = $(temp_HI))");
    plt.xlabel("MC Sweeps"); plt.ylabel("Var Delta Energy"); plt.legend(loc = "upper left"); 
    plt.title("SEQ $(i)"); plt.savefig("../figures/DBD_ayse_trial/var_en_SEQ_$(i).png") end



dist_AW = []; var_dist_AW = []; dist_HI = []; var_dist_HI = []; dist_LO = []; var_dist_LO = [];
for i in 1:3
    dist = []; var_dist = [];
    for t in 1:N_points_AW
        msa_dist = ham_dist(start_msa[:,i], res_all_AW[i].step_msa[t]) ./ L
        push!(dist, mean(msa_dist)[1])
        push!(var_dist, var(msa_dist)[1])
    end
    push!(dist_AW, dist)
    push!(var_dist_AW, var_dist)
    dist = []; var_dist = [];
    for t in 1:N_points_HI
        msa_dist = ham_dist(start_msa[:,i], res_all_HI[i].step_msa[t]) ./ L
        push!(dist, mean(msa_dist)[1])
        push!(var_dist, var(msa_dist)[1])
    end
    push!(dist_HI, dist)
    push!(var_dist_HI, var_dist)
    dist = []; var_dist = [];
    for t in 1:N_points_LO
        msa_dist = ham_dist(start_msa[:,i], res_all_LO[i].step_msa[t]) ./ L
        push!(dist, mean(msa_dist)[1])
        push!(var_dist, var(msa_dist)[1])
    end
    push!(dist_LO, dist)
    push!(var_dist_LO, var_dist)
end

steps_LO = res_all_LO[1].steps ./ L; steps_HI = res_all_HI[1].steps ./ L; steps_AW = (2 .- reverse(res_all_AW[1].steps)) ./ L;
for i in 1:3
    close("all"); plt.errorbar(steps_AW, dist_AW[i], yerr = sqrt.(var_dist_AW[i]), marker = "o", 
        label = "AW (T = $(temp_AW))");
    plt.errorbar(steps_LO, dist_LO[i], yerr = sqrt.(var_dist_LO[i]), marker = "o", label = "LO (T = $(temp_LO))");
    plt.errorbar(steps_HI, dist_HI[i], yerr = sqrt.(var_dist_HI[i]), marker = "o", label = "HI (T = $(temp_HI))");
    plt.xlabel("MC Sweeps"); plt.ylabel("Hamming Distance"); plt.legend(loc = "upper left"); 
    plt.title("SEQ $(i)"); plt.savefig("../figures/DBD_ayse_trial/dist_SEQ_$(i).png") 
end

for i in 1:3
    close("all"); plt.plot(steps_AW, var_dist_AW[i], marker = "o", label = "AW (T = $(temp_AW))");
    plt.plot(steps_LO, var_dist_LO[i], marker = "o", label = "LO (T = $(temp_LO))");
    plt.plot(steps_HI, var_dist_HI[i], marker = "o", label = "HI (T = $(temp_HI))");
    plt.xlabel("MC Sweeps"); plt.ylabel("Var Hamming Distance"); plt.legend(loc = "upper left"); 
    plt.title("SEQ $(i)"); plt.savefig("../figures/DBD_ayse_trial/var_dist_SEQ_$(i).png") 
end


```

## Try adaptive walk by incrementing temperature to reach saturation

```julia
using Revise, Genie, DCAUtils, JLD2, PyPlot, Statistics


nat_msa  = read_fasta_alignment("../Gen.jl/data/alignments/natural/DBD_alignment.uniref90.cov80.a2m", 0.9);
w = compute_weights(nat_msa, 22, 0.2)[1];
@load "../data_Genie/pars_dbd.jld2"; h = h_dbd; J = J_dbd;
@load"../data_Anc/3seqs_data_for_tree_sampling.jld2";

start_ens = energy(start_msa, h, J);

q = 21; L =size(nat_msa,1);

#just one
temp_HI = 0.2; temp_LO = 1.2; step_msa_all_AW = []; res_all_HI = []; res_all_LO = [];
N_steps_HI = 500; N_steps_LO = 500; each_step = 50; N_seqs = 1000; p = 0.; 

small_steps = 20; 
temps = [0.6/(1.3^(i-1)) for i in 1:10];
steps_HI = [];

for i in 1:3
    step_msa = []
    push!(step_msa, hcat([start_msa[:,i] for _ in 1:N_seqs]...))
    res = run_evolution(hcat([start_msa[:,i] for _ in 1:N_seqs]...), h, J, temp = temps[1],
        N_steps = small_steps, p = p);
    push!(step_msa, res.msa)
    for temp in temps[2:end]
        res = run_evolution(step_msa[end], h, J, temp = temp, N_steps = small_steps, p = p);
        push!(step_msa, res.msa)
    end
    push!(step_msa_all_AW, step_msa)
    
    init = step_msa_all_AW[i][end]
    @time res_HI = run_evolution(init, h, J, temp = temp_HI,
    N_steps = N_steps_HI, each_step = each_step, p = p);
    push!(res_all_HI, res_HI)
    @time res_LO = run_evolution(init, h, J, temp = temp_LO,
    N_steps = N_steps_LO, each_step = each_step, p = p);
    push!(res_all_LO, res_LO)
end
    
    

N_points_AW = length(step_msa_all_AW[1]);
N_points_LO = length(res_all_LO[1].step_msa); 
N_points_HI = length(res_all_HI[1].step_msa); 

en_AW = []; var_en_AW = []; en_HI = []; var_en_HI = []; en_LO = []; var_en_LO = [];
for i in 1:3
    en = []; var_en = [];
    for t in 1:N_points_AW
        msa_en = energy(step_msa_all_AW[i][t], h, J) .- start_ens[i]
        push!(en, mean(msa_en)[1])
        push!(var_en, var(msa_en)[1])
    end
    push!(en_AW, en)
    push!(var_en_AW, var_en)
    
    en = []; var_en = [];
    for t in 1:N_points_HI
        msa_en = energy(res_all_HI[i].step_msa[t], h, J) .- start_ens[i]
        push!(en, mean(msa_en)[1])
        push!(var_en, var(msa_en)[1])
    end
    push!(en_HI, en)
    push!(var_en_HI, var_en)
    
    en = []; var_en = [];
    for t in 1:N_points_LO
        msa_en = energy(res_all_LO[i].step_msa[t], h, J) .- start_ens[i]
        push!(en, mean(msa_en)[1])
        push!(var_en, var(msa_en)[1])
    end
    push!(en_LO, en)
    push!(var_en_LO, var_en)
end

steps_AW = reverse([-small_steps*i for i in 0:10]) ./L;
steps_LO = res_all_LO[1].steps ./ L; steps_HI = res_all_HI[1].steps ./ L; 

for i in 1:3
close("all"); plt.errorbar(steps_AW, .- en_AW[i], yerr = sqrt.(var_en_AW[i]), marker = "o", label = "AW (final T = $(round(temps[end], digits = 2)))");
    plt.errorbar(steps_LO, .- en_LO[i], yerr = sqrt.(var_en_LO[i]), marker = "o", label = "LO (T = $(temp_LO))");
    plt.errorbar(steps_HI, .- en_HI[i], yerr = sqrt.(var_en_HI[i]), marker = "o", label = "HI (T = $(temp_HI))");
    plt.xlabel("MC Sweeps"); plt.ylabel(" - Delta Energy"); plt.legend(loc = "upper left"); 
    plt.title("SEQ $(i)"); plt.savefig("../figures/DBD_ayse_trial/adap_walk/en_SEQ_$(i).png") end

for i in 1:3
close("all"); plt.plot(steps_AW, var_en_AW[i], marker = "o", label = "AW (final T = $(round(temps[end], digits = 2)))");
    plt.plot(steps_LO, var_en_LO[i], marker = "o", label = "LO (T = $(temp_LO))");
    plt.plot(steps_HI, var_en_HI[i], marker = "o", label = "HI (T = $(temp_HI))");
    plt.xlabel("MC Sweeps"); plt.ylabel("Var Delta Energy"); plt.legend(loc = "upper left"); 
    plt.title("SEQ $(i)"); plt.savefig("../figures/DBD_ayse_trial/adap_walk/var_en_SEQ_$(i).png") end

```
