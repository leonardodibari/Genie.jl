```julia
using PyPlot

# Your data matrix
data = [
    fantini_mcmc[1].corr      fantini_mcmc[2].corr      stiffler_mcmc[1].corr      stiffler_mcmc[2].corr      aac_mcmc.corr;
    fantini_popgen[1].corr    fantini_popgen[2].corr    stiffler_popgen[1].corr    stiffler_popgen[2].corr    aac_popgen.corr;
    fantini_mcmc_tree[1].corr fantini_mcmc_tree[2].corr stiffler_mcmc_tree[1].corr stiffler_mcmc_tree[2].corr aac_mcmc_tree.corr
]

# Optional: convert values to strings with limited decimal places
cell_text = [string(round(val, digits=3)) for val in data]

# Reshape to 3x5 string matrix
cell_text = reshape(cell_text, size(data)...)

# Define row and column labels
row_labels = ["MCMC", "PopGen", "MCMC+Tree"]
col_labels = ["Fantini 1", "Fantini 2", "Stiffler 1", "Stiffler 2", "AAC"]

# Create the table
fig, ax = subplots()
ax.axis("off")  # Hide axes

table(cellText=cell_text, rowLabels=row_labels, colLabels=col_labels, loc="center")

tight_layout()
show()
```
